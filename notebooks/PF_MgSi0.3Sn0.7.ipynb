```python
import sys
sys.path.append('../')
import spb_scattering_mech as spb
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_eta = 100 #number of doping levels
pf = np.zeros(n_eta) #initialize power factor array
s = 0.5
muW = 392E-4 #m^2/Vs
T = 300

i = 0
eta_list = np.linspace(-4, 10, n_eta)
```

```python
for eta in eta_list:
    S = spb.seebeck_s(eta, l = 0)
    sigma = spb.conductivity_s_simp(eta, T, muW, l = 0)
    pf[i] = S**2 * sigma
    i = i+1
```

```python
plt.plot(eta_list, pf)
```

```python
muW = 141.5E-4
pf2 = np.zeros(n_eta)
j = 0

for eta in eta_list:
    S = spb.seebeck_s(eta, l = 0.5)
    sigma = spb.conductivity_s_simp(eta, T, muW, l = 0.5)
    pf2[j] = S**2 * sigma
    j = j+1
```

```python
plt.plot(eta_list, pf2)
```
